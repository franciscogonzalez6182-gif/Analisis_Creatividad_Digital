```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar el dataset principal
df = pd.read_csv('datos/campanas_creatividad_digital.csv')

# 2. Dimensiones (Filas, Columnas)
print(f"Dimensiones: {df.shape}")

# 3. Tipos de datos y nulos
print("\nInformación de columnas:")
print(df.info())

# 4. Cantidad de valores nulos por columna
print("\nValores nulos encontrados:")
print(df.isna().sum())

# 5. Identificar columnas numéricas y categóricas
columnas_numericas = df.select_dtypes(include=['number']).columns.tolist()
columnas_categoricas = df.select_dtypes(include=['object']).columns.tolist()

# 6. Mostrar las primeras 5 filas
print("\nPrimeras 5 filas:")
display(df.head(5))

# 7. Mostrar las últimas 5 filas
print("\nÚltimas 5 filas:")
display(df.tail(5))
```

## Indicadores derivados
 Los indicadores se calculan a partir de las columnas originales. Van de simples (una operacion) a compuestos (varias columnas combinadas).
* **`CTR`**: Indica qué tan atractivo es el anuncio visualmente.
* **`CVR`**: Indica si la oferta o página de destino realmente convence al usuario de comprar o registrarse.
* **`ROI`**: Es la métrica reina. Si es mayor a `0`, la campaña generó más dinero del que costó.
* **`CPC`**: mide el costo unitario que pagas por cada vez que un usuario hace clic en tu anuncio.
* **`CPM`**: mide el costo de mostrar tu anuncio mil veces, independientemente de si los usuarios hacen clic o no.

```python
# 1. Indicadores básicos
df['ctr'] = df['clicks'] / df['impresiones']  # Tasa de clics
df['cvr'] = df['conversiones'] / df['clicks']  # Tasa de conversión

# 2. Indicadores de costo
df['cpc'] = df['costo_usd'] / df['clicks']  # Costo por click
df['cpm'] = (df['costo_usd'] / df['impresiones']) * 1000  # Costo por mil impresiones

# 3. Indicador compuesto
# El ROI mide la rentabilidad: (Ganancia Neta / Inversión)
df['roi'] = (df['ingresos_usd'] - df['costo_usd']) / df['costo_usd']

# Verificamos que no haya valores nulos (por si hubo divisiones entre cero)
print("Valores nulos en nuevas métricas:")
print(df[['ctr', 'cvr', 'cpc', 'cpm', 'roi']].isna().sum())

# Mostramos las primeras filas con los nuevos cálculos
df[['id_campana', 'plataforma', 'ctr', 'cvr', 'cpc', 'cpm','roi']].head()
```

## Preguntas de negocio con filtros

```python
# --- PARTE 3: PREGUNTAS DE NEGOCIO CON FILTROS ---

# q1 — Campañas con ROI negativo y costo por encima de la mediana
mediana_costo = df['costo_usd'].median()
q1 = df[(df['roi'] < 0) & (df['costo_usd'] > mediana_costo)]

print(f"q1 - Cantidad de campañas con ROI negativo y costo alto: {len(q1)}")
print("\nPlataformas que concentran más casos de ROI negativo (Costo > Mediana):")
print(q1['plataforma'].value_counts())

# q2 — Top 10 campañas por ROI únicamente en TikTok e Instagram
filtro_plataformas = df['plataforma'].isin(['TikTok', 'Instagram'])
q2 = df[filtro_plataformas].sort_values(by='roi', ascending=False).head(10)

print("\nq2 - Top 10 campañas por ROI (TikTok e Instagram):")
display(q2[['id_campana', 'plataforma', 'formato', 'roi']])

# q3 — Formato con el CTR promedio más alto
q3 = df.groupby('formato')[['ctr']].mean().sort_values(by='ctr', ascending=False)

print("\nq3 - Formato con el CTR promedio más alto (ordenado de mayor a menor):")
display(q3)

# q4 — País con el mayor ingreso total
q4 = df.groupby('pais')[['ingresos_usd']].sum().sort_values(by='ingresos_usd', ascending=False)
pais_lider = q4.index[0]
ingreso_max = q4.iloc[0, 0]

print(f"\nq4 - El país con el mayor ingreso total es {pais_lider} con ${ingreso_max:,.2f}")
```

## Estadistica descriptiva: dispersion 
(qué tan alejados están los datos de sus promedios. Esto es vital para saber si tus resultados de marketing son constantes o si tienes campañas con resultados muy "extremos" (altos o bajos).)

```python
# --- PARTE 5: ESTADÍSTICA DESCRIPTIVA: DISPERSIÓN ---

# 1. Definimos las columnas numéricas solicitadas
cols_numericas = [
    'costo_usd', 'impresiones', 'clicks', 'conversiones', 
    'ingresos_usd', 'ctr', 'cvr', 'cpc', 'cpm', 'roi'
]

# 2. Realizamos los cálculos estadísticos
desviacion = df[cols_numericas].std()    # Desviación estándar
varianza = df[cols_numericas].var()      # Varianza
rango = df[cols_numericas].max() - df[cols_numericas].min() # Rango (Máximo - Mínimo)

# 3. Calculamos el Rango Intercuartílico (IQR)
Q1 = df[cols_numericas].quantile(0.25)   # Primer cuartil (25%)
Q3 = df[cols_numericas].quantile(0.75)   # Tercer cuartil (75%)
iqr = Q3 - Q1                           # Amplitud del 50% central

# 4. Presentamos los resultados en el dataframe 'dispersion'
dispersion = pd.DataFrame({
    'Desviación Estándar': desviacion,
    'Varianza': varianza,
    'Rango': rango,
    'IQR': iqr
})

print("Dataframe: dispersion")
display(dispersion)
```

### Interpretación de la Variabilidad:
La columna con mayor variabilidad relativa suele ser `roi` o `ingresos_usd`. 

Esto se debe a que, en el marketing digital, el rendimiento no es lineal: una campaña "viral" puede generar ingresos masivos (miles de dólares) mientras que otra fallida puede generar casi cero. El Rango del ROI es particularmente amplio porque incluye desde pérdidas casi totales (-0.95 o -95%) hasta retornos de inversión extremadamente altos (superiores a 20 o 30).

## Deteccion de outliers con IQR 

```python
# --- PARTE 6: DETECCIÓN DE OUTLIERS CON IQR ---

def detectar_outliers(data, columna):
    # 1. Calculamos Q1 (25%) y Q3 (75%)
    q1 = data[columna].quantile(0.25)
    q3 = data[columna].quantile(0.75)
    
    # 2. Calculamos el IQR
    iqr = q3 - q1
    
    # 3. Definimos los límites
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    # 4. Filtramos las filas fuera de los límites
    outliers = data[(data[columna] < limite_inferior) | (data[columna] > limite_superior)]
    
    return outliers, limite_inferior, limite_superior

# Aplicar para CTR
outliers_ctr, lim_inf_ctr, lim_sup_ctr = detectar_outliers(df, 'ctr')
print(f"Outliers detectados en CTR: {len(outliers_ctr)}")
display(outliers_ctr[['id_campana', 'plataforma', 'formato', 'ctr']])

# Aplicar para ROI
outliers_roi, lim_inf_roi, lim_sup_roi = detectar_outliers(df, 'roi')
print(f"\nOutliers detectados en ROI: {len(outliers_roi)}")
display(outliers_roi[['id_campana', 'plataforma', 'formato', 'roi']])
```

### Análisis de Outliers:

* ¿Cuántos outliers hay?: 
> En el CTR se detectaron 3 campañas atípicas, mientras que en el ROI fueron 12.

* ¿Son valores altos o bajos?: 
> En ambos casos, se trata de valores altos.

>El límite superior de CTR fue superado por campañas con valores superiores al 3.1% (0.031).

>El ROI muestra valores extremos que llegan hasta 33.42, lo que significa que esas campañas generaron 33 veces más de lo que costaron.

* ¿Qué plataformas aparecen más?: TikTok domina absolutamente la lista de outliers.

>En CTR, el 100% de los casos (3 de 3) pertenecen a TikTok con el formato Reel.

>En ROI, TikTok concentra 10 de los 12 casos detectados, abarcando diversos formatos como Story, Reel, Carrusel e incluso Video Largo.

>Otras plataformas con presencia mínima en los outliers de ROI son YouTube (1 caso) e Instagram (1 caso).

# Merge con tabla de clientes

```python
# --- PARTE 8: MERGE CON TABLA DE CLIENTES ---

# 1. Carga del archivo de clientes
clientes = pd.read_csv('datos/clientes.csv')

# m1 — Left merge (Conserva todas las campañas)
m1 = pd.merge(df, clientes, on='cliente', how='left')
print(f"Filas en df original: {len(df)}")
print(f"Filas en m1 (Left): {len(m1)}")
# Verificación: m1 debe tener 180 filas (igual que df)

# m2 — Inner merge (Solo donde hay coincidencia en ambos)
m2 = pd.merge(df, clientes, on='cliente', how='inner')
print(f"Filas en m2 (Inner): {len(m2)}")

# m3 — Outer merge (Todos los registros de ambas tablas)
m3 = pd.merge(df, clientes, on='cliente', how='outer')

# Identificar clientes sin campañas en m3
clientes_sin_campana = m3[m3['id_campana'].isna()]['cliente'].unique()
print(f"Clientes sin campañas registradas: {clientes_sin_campana}")

# --- ANALISIS POST-MERGE ---

# pm1 — ROI promedio agrupado por industria
pm1 = m1.groupby('industria')['roi'].mean().sort_values(ascending=False)
print("\npm1 - ROI promedio por industria:")
display(pm1)

# pm2 — CTR promedio agrupado por nivel_contrato
pm2 = m1.groupby('nivel_contrato')['ctr'].mean().sort_values(ascending=False)
print("\npm2 - CTR promedio por nivel de contrato:")
display(pm2)

# pm3 — Campañas de clientes grandes con ROI positivo
clientes_grandes = m1[m1['tamano'] == 'Grande']
grandes_roi_pos = clientes_grandes[clientes_grandes['roi'] > 0]
num_grandes_roi_pos = len(grandes_roi_pos)
pct_grandes_roi_pos = (num_grandes_roi_pos / len(clientes_grandes)) * 100

print(f"\npm3 - Clientes Grandes con ROI positivo: {num_grandes_roi_pos}")
print(f"Representan el {pct_grandes_roi_pos:.2f}% de las campañas de clientes grandes.")
```

### Diferencia entre m1 (`Left`) y m2 (`Inner`)
* **m1 (Left Merge)**: Mantiene todas las `180` filas de tus campañas. Si un cliente no está en la tabla `clientes.csv`, verás valores `NaN` en las columnas nuevas. Esto es útil para no perder datos de inversión.

* **m2 (Inner Merge)**: Solo conserva las filas donde el cliente existe en ambas tablas. Si `m2` tiene menos de `180` filas, significa que hay campañas de clientes "desconocidos" o que no están en tu base de datos maestra de clientes.

### Interpretación de Resultados:
* **pm1 (Industria)**: El resultado te dirá qué sector (ej. `Tecnología` o `Moda`) es el más rentable. Según el ejemplo del reporte, podrías esperar que las industrias con audiencias jóvenes tengan mejor desempeño en plataformas como `TikTok`.

* **pm2 (Nivel de Contrato)**: Aquí verificarás si los clientes `Enterprise` (que suelen pagar más por servicios Premium) obtienen realmente un mejor `CTR` que los clientes con contrato Básico.

* **pm3 (Eficiencia en Grandes Clientes)**: Esta métrica es vital para la cuenta de resultados, ya que los clientes grandes suelen manejar los presupuestos más altos. Un porcentaje alto aquí indica una gestión saludable de las cuentas clave.
